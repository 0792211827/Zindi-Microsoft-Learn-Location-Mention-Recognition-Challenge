```python
# Removing Digits
train['text'] = train['text'].apply(lambda d: re.sub(r'\d','', d) if isinstance(d, str) else d)
test['text'] = test['text'].apply(lambda d: re.sub(r'\d','', d) if isinstance(d, str) else d)
```

```python
#removing commas
train['text'] = train['text'].apply(lambda v: re.sub(',', '', v) if isinstance(v, str) else v)
test['text'] = test['text'].apply(lambda v: re.sub(',', '', v) if isinstance(v, str) else v)
```

```python
#a function to remove emojis
def remove_emoji(text):
    if not isinstance(text, str):
        return text
    emoji_pattern = re.compile(
        "["
        "\U0001F600-\U0001F64F"  # Emoticons
        "\U0001F300-\U0001F5FF"  # Symbols & Pictographs
        "\U0001F680-\U0001F6FF"  # Transport & Map Symbols
        "\U0001F700-\U0001F77F"  # Alchemical Symbols
        "\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
        "\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
        "\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
        "\U0001FA00-\U0001FA6F"  # Chess Symbols
        "\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
        "\U00002702-\U000027B0"  # Dingbats
        "\U000024C2-\U0001F251"  # Enclosed Characters
        "]+", flags=re.UNICODE)

    return emoji_pattern.sub(r'', text)

train['text'] = train['text'].apply(remove_emoji)
test['text'] = test['text'].apply(remove_emoji)
```

```python
#remove HTML tags
def remove_html(text):
    if not isinstance(text, str):
        return text
    html_pattern = re.compile(r'<.*?>')
    return html_pattern.sub(r'', text)
train['text'] = train['text'].apply(remove_html)
test['text'] = test['text'].apply(remove_html)
```

```python
# remove URLs
def remove_URL(text):
    if not isinstance(text, str):
        return text
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)

train['text'] = train['text'].apply(remove_URL)
test['text'] = test['text'].apply(remove_URL)
```

```python
# Initialize an empty list to store the extracted locations for each row
extracted_locations = []

# Process all rows in the DataFrame
for index, row in train.iterrows():
     # Ensure the cleaned_text is a string
    text = row['text'] if isinstance(row['text'], str) else ''
    temp_dict['text'] = text
    temp_dict['locations'] = []

    # Initialize an empty list to store locations for the current row
    row_locations = []

    # Split the location column into individual locations if applicable
    locations = row['location'].split() if isinstance(row['location'], str) else []

    # Use regular expressions to find each location in the text
    for location in locations:
        pattern = re.compile(re.escape(location), re.IGNORECASE)
        match = pattern.search(text)

        # If a location is found, store the start and end positions
        if match:
            start = match.start()
            end = match.end()
            row_locations.append((start, end, 'LOCATION'))

    # Append the extracted locations for the current row to the list
    extracted_locations.append(row_locations)

# Create a new column 'extracted_locations' in the DataFrame
train['extracted_locations'] = extracted_locations

# Print the first few rows to verify
train[['text', 'location', 'extracted_locations']].head()
```

```python
import spacy
from spacy.tokens import DocBin
from tqdm import tqdm

# Load a new blank spaCy model for English
nlp = spacy.blank("en")

# Initialize DocBin for storing processed documents
doc_bin = DocBin()
```

```python
from tqdm import tqdm
from spacy.util import filter_spans

# Use the train DataFrame directly as the training data
training_data = train  # Ensure you're using the DataFrame with the 'extracted_locations' column

for training_example in tqdm(training_data.itertuples(), total=len(training_data)):
    text = training_example.text  # Use the 'text' column from the DataFrame
    labels = training_example.extracted_locations  # Use the 'extracted_locations' column

    # Create a doc object from the text
    doc = nlp.make_doc(text)
    ents = []

    # Add entities to the doc
    for start, end, label in labels:
        span = doc.char_span(start, end, label=label, alignment_mode="contract")
        if span is None:
            print(f"Skipping entity: {(start, end, label)}")
        else:
            ents.append(span)

    # Filter out overlapping spans
    filtered_ents = filter_spans(ents)
    doc.ents = filtered_ents

    # Add the doc to the DocBin
    doc_bin.add(doc)

# Save the processed data to a file
doc_bin.to_disk("train.spacy")

```
